1. Tentukan Library yang di gunakan

```python
import numpy as np
import pandas as pd 
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
```

MEMUAT DATASET

```python
# Langkah 1: Memuat data dari file CSV
DataGizi = pd.read_csv('DataGizi.csv')
array = DataGizi.values
```

```python
DataGizi.head()
```

```python
DataGizi.shape
```

```python
DataGizi['Label'].value_counts()
```

```python
#memuat data labelencoder
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
for column in DataGizi:
    if DataGizi[column].dtypes ==object:
        DataGizi[column] = le.fit_transform(DataGizi[column])
```

```python
for column in DataGizi:
    if DataGizi[column].dtypes ==float:
        DataGizi[column] = le.fit_transform(DataGizi[column])
```

```python
DataGizi.info()
```

```python
DataGizi.head()
```

```python
#data yang telah di labelkan 
DataGizi['Label'].value_counts()
```

```python
#memisahkan data dan label

x = DataGizi.drop(columns='Label', axis=1)
y = DataGizi['Label']
```

```python
print (x)
```

```python
print(y)
```

UNTUK MENGKONVERSI DATA YANG SUDAH DI NORMALISASIKAN

3. STANDARISASI DATA

```python
scaler = StandardScaler()
```

```python
scaler.fit(x)
```

```python
standarized_data = scaler.transform(x)
```

```python
print(standarized_data)
```

```python
x = standarized_data
y = DataGizi['Label']
```

```python
print(x)
print(y)
```

4. MEMISAHKAN DATA TESTING DAN DATA TRANING

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, stratify=y, random_state=2)
```

```python
print(x.shape, x_train.shape, x_test.shape)
```

5. MEMBUAT DATA LATIH MENGGUNAKAN NAIVE BAYES

```python
from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()
```

```python
classifier.fit(x_train, y_train)
```

6. MEMBUAT MODEL EVALUASI UNTUK MENGUKUR TINGKAT AKURASI

```python
x_train_prediction = classifier.predict(x_train)
traning_data_accuracy = accuracy_score(x_train_prediction, y_train)
```

```python
print('Akurasi Data Traning Adalah =', traning_data_accuracy)
```

```python
x_test_prediction = classifier.predict(x_test)
test_data_accuracy = accuracy_score(x_test_prediction, y_test)
```

```python
print('Akurasi Data Testing Adalah =', test_data_accuracy)
```

7. MEMBUTA MODEL PREDIKSI

```python
input_data=(280, 1,	1.8, 46.0, 90, 6.2, 69.0)

input_data_as_numpy_array = np.array(input_data)

input_data_reshape = input_data_as_numpy_array.reshape(1,-1)

std_data = scaler.transform(input_data_reshape)
print(std_data)

prediction = classifier.predict(std_data)
print(prediction)

if (prediction[0] == 0):
    print('pasien tidak berisiko gizi lebih')
else :
    print('pasien masuk gizi kurang')
```
